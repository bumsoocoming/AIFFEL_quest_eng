```python
import math # 제곱 연산을 위해 가져오기

# 프로그램 전체를 반복하기 위한 무한 루프
while True:
    # 1. 첫 번째 숫자 입력받기
    while True:
        num1_input = input("첫 번째 정수를 입력하세요: ")
        try:
            num1 = int(num1_input)
            break # 정수 입력에 성공하면 반복문 탈출
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 2. 두 번째 숫자 입력받기
    while True:
        num2_input = input("두 번째 정수를 입력하세요: ")
        try:
            num2 = int(num2_input)
            break # 정수 입력에 성공하면 반복문 탈출
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 3. 연산자 입력받기
    operator = input("연산자를 입력하세요 (+, -, *, /, **): ")

    # 4. 실제 계산하기
    if operator == "+":
        print("결과:", num1 + num2)
    elif operator == "-":
        print("결과:", num1 - num2)
    elif operator == "*":
        print("결과:", num1 * num2)
    elif operator == "/":
        try:
            print("결과:", num1 / num2)
        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다.")
    elif operator == "**":
        # math 모듈을 사용하라는 조건 반영
        result = math.pow(num1, num2)
        print("결과:", result)
    else:
        print("오류: 지원하지 않는 연산자입니다.")

    # 5. 더 할지 물어보기
    go_on = input("계속 하시겠습니까? (y/n): ")
    if go_on == "n":
        print("계산기를 종료합니다.")
        break # 가장 바깥쪽 while문 탈출

```
