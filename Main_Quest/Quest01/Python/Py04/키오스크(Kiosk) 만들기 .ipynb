# 사각형 넓이 구하기

```python
# 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
        self.square = int(input(
            '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n'
            '1.직사각형 2.평행사변형 3.사다리꼴 \n >>> '
        ))

        if self.square == 1:
            print(self.rect())

        elif self.square == 2:
            print(self.par())

        elif self.square == 3:
            print(self.trape())

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input(
            '가로, 세로를 입력하세요. 예시 : 5,10\n >>> '
        ).split(','))
        area = width * vertical
        return f'직사각형의 넓이는 : {area}'

    def par(self):
        base, height = map(int, input(
            '밑변, 높이를 입력하세요. 예시 : 5,10\n >>> '
        ).split(','))
        area = base * height
        return f'평행사변형의 넓이는 : {area}'

    def trape(self):
        base, top, height = map(int, input(
            '밑변, 윗변, 높이를 입력하세요. 예시 : 10,6,5\n >>> '
        ).split(','))
        area = (base + top) * height / 2
        return f'사다리꼴의 넓이는 : {area}'


# 실행
a = Square()
```

# kiosk

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []
        self.order_price = []

    def menu_print(self):
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")

    def menu_select(self):
        while True:
            n = int(input("\n음료 번호(종료는 0): "))
            if n == 0: break
            if 1 <= n <= len(self.menu):
                t = int(input("HOT(1) / ICE(2): "))
                self.temp = "HOT" if t == 1 else "ICE"
                self.order_menu.append(f"{self.temp} {self.menu[n-1]}")
                self.order_price.append(self.price[n-1])
            else:
                print("잘못된 번호입니다.")

    def pay(self):
        print(f"\n총액: {sum(self.order_price)}원")
        p = int(input("현금(1) / 카드(2): "))
        print("결제가 진행됩니다...")

    def table(self):
        print("-" * 20)
        for i in range(len(self.order_menu)):
            print(f"{self.order_menu[i]} : {self.order_price[i]}원")
        print(f"합계: {sum(self.order_price)}원")
        print("-" * 20)

# 실행부
cafe = Kiosk()
cafe.menu_print()
cafe.menu_select()
cafe.pay()
cafe.table()
```
