```python
# Q4. 입금 메서드
import random

class Account:

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)

        self.account_number = f"{num1}-{num2}-{num3}"


    # 입금 메서드
    def deposit(self, amount):
        if amount >= 1:  # 1원 이상만 가능
            self.balance += amount
        else:
            print("1원 이상만 입금 가능")


# 테스트
a1 = Account("김범수", 10000)

a1.deposit(5000)
print(a1.balance)  # 15000

a1.deposit(0)  # 에러 메시지
```

```python
# Q5. 출금 메서드
import random

class Account:

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)

        self.account_number = f"{num1}-{num2}-{num3}"


    # 출금 메서드
    def withdraw(self, amount):
        if amount <= self.balance:  # 잔액보다 작거나 같으면 가능
            self.balance -= amount
        else:
            print("잔액 부족")


# 테스트
a1 = Account("김범수", 10000)

a1.withdraw(3000)
print(a1.balance)  # 7000

a1.withdraw(20000)  # 잔액 부족
```

```python
# Q7. 이자 지급하기
import random

class Account:

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        self.deposit_count = 0  # 입금 횟수

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)

        self.account_number = f"{num1}-{num2}-{num3}"


    # 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1  # 입금 횟수 +1

            # 5번마다 이자 지급
            if self.deposit_count % 5 == 0:
                self.balance += self.balance * 0.01  # 1% 이자
        else:
            print("1원 이상만 입금 가능")


# 테스트
a1 = Account("김범수", 10000)

a1.deposit(1000)
a1.deposit(1000)
a1.deposit(1000)
a1.deposit(1000)
a1.deposit(1000)  # 여기서 이자 발생

print(a1.balance)
```

```python
import random

class Account:

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)

        self.account_number = f"{num1}-{num2}-{num3}"

        # ✅ 입금 / 출금 기록 저장 리스트
        self.deposit_list = []
        self.withdraw_list = []

    # ✅ 입금 메서드
    def deposit(self, amount):
        self.balance += amount
        self.deposit_list.append(amount)

    # ✅ 출금 메서드
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.withdraw_list.append(amount)
        else:
            print("잔액이 부족합니다.")

    # ✅ 입금 내역 출력
    def deposit_history(self):
        print("입금 내역:", self.deposit_list)

    # ✅ 출금 내역 출력
    def withdraw_history(self):
        print("출금 내역:", self.withdraw_list)


# 계좌 생성
a1 = Account("김범수", 500000)

# 테스트
a1.deposit(100000)
a1.deposit(200000)
a1.withdraw(150000)

a1.deposit_history()
a1.withdraw_history()
```
